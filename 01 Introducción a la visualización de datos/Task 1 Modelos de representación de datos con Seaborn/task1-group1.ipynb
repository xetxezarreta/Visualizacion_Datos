```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white")

df1 = pd.read_csv("dataset-airport-1.csv")
df1.head()
```

# Numero de pasajeros hombres en funcion del tiempo

```python
plt.figure(figsize=(16, 6))
ax2 = sns.lineplot(x="Datetime", y="Male", data=df1)

#ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45, horizontalalignment='right')
ax2.get_xticklabels()[1]
```

```python
df2 = pd.read_csv("dataset-airport-2.csv")
df2.head()
```

# Distribucion de los pasageros por aerolinea en el tiempo

```python
plt.figure(figsize=(16, 6))

chart = sns.catplot(x="Passengers", y="year", hue="Airline", data=df2)
```

# Numero de pasajeros por aerolinea (media)

```python
plt.figure(figsize=(16, 6))
ax = sns.barplot(x="Airline", y="Passengers", data=df2, ci=None,  palette=sns.color_palette("RdBu", n_colors=8))
```
