### Read datasets

```python
import pandas as pd
```

```python
df1 = pd.read_csv('datasets/dataset-airport-1.csv')
df1.head()
```

```python
df2 = pd.read_csv('datasets/dataset-airport-2.csv')
df2.head()
```

```python
df3 = pd.read_csv('datasets/dataset-airport-3.csv')
df3.head()
```

### Seaborn

##### Chart 1

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(26, 6))
sns.set(style="darkgrid")
ax = sns.lineplot(x="Datetime", y="Passengers",data=df1, palette=['#ff0011'])
ax = ax.set_xticklabels([pd.to_datetime(tm, unit='ms').strftime('%Y-%m-%d') for tm in ax.get_xticks()], rotation=50)
```

Evolución de pasajeros por dias. 

##### Chart 2

```python
df3['datetime'] = pd.to_datetime(df3['datetime'])

chart2_df = df3.groupby([df3['datetime'].dt.year]).mean()
chart2_df.head()
```

```python
sns.barplot(x=chart2_df.index, y='horas vuelo', data=chart2_df)
```

La media de horas de vuelo por cada año en forma de barras.

No se aprecia mucha diferencia.

##### Chart 3

```python
plt.figure(figsize=(24, 8))
sns.set(style="darkgrid")
ax = sns.lineplot(x='year', y='value', hue='variable', 
                  data=pd.melt(df2, id_vars=['year'], value_vars=['Bussiness', 'First Class']), ci=None)
```

Evolución de los pasajeros en Bussiness y First Class de todas las compañías por año.

##### Chart 4

```python
plt.figure(figsize=(24, 8))
sns.barplot(x='Airline', y='Passengers', hue='year', data=df2, ci=None)
```

Número de pasajeros en cada aerolinea por cada año.
