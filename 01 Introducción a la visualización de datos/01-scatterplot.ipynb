```python
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

### Indian diabetes

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

### Scatterplot 1

```python
sns.scatterplot(x="Glucose", y="Age", data=df, hue='Outcome')
```

REPRESENTADO: 
- Relación entre la edad y el nivel de glucosa y su influencia en la variable de salida (1=diabético, 0=no diabético)

CONCLUSIONES:
- La edad no afecta mucho al nivel de glucosa.
- La edad no afecta a ser diabético o no.
- Mayor nivel de glucosa, mas probabilidad de ser diabético.

### Scatterplot 2

```python
sns.scatterplot(x="BloodPressure", y="Glucose", data=df, hue='Outcome')
```

REPRESENTADO: 
- Relación entre la presión sanguinea y el nivel de glucosa y su influencia en la variable de salida (1=diabético, 0=no diabético)

CONCLUSIONES:
- Hay outliers (ej: cuando la presión sanguinea es 0)
- El nivel de glucosa no afecta a la presión sanguinea.
- Mayor nivel de glucosa, mas probabilidad de ser diabético.
- La presión sanguinea no afecta a ser diabético o no.
