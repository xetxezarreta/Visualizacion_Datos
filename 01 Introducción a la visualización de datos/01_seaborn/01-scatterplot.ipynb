```python
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

### Indian diabetes

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

### Scatterplot 1

```python
sns.scatterplot(x="Glucose", y="Age", data=df, hue='Outcome')
```

REPRESENTADO: 
- Relacin entre la edad y el nivel de glucosa y su influencia en la variable de salida (1=diabtico, 0=no diabtico)

CONCLUSIONES:
- La edad no afecta mucho al nivel de glucosa.
- La edad no afecta a ser diabtico o no.
- Mayor nivel de glucosa, mas probabilidad de ser diabtico.

### Scatterplot 2

```python
sns.scatterplot(x="BloodPressure", y="Glucose", data=df, hue='Outcome')
```

REPRESENTADO: 
- Relacin entre la presin sanguinea y el nivel de glucosa y su influencia en la variable de salida (1=diabtico, 0=no diabtico)

CONCLUSIONES:
- Hay outliers (ej: cuando la presin sanguinea es 0)
- El nivel de glucosa no afecta a la presin sanguinea.
- Mayor nivel de glucosa, mas probabilidad de ser diabtico.
- La presin sanguinea no afecta a ser diabtico o no.
